### Sensibilidades en T y r simultaneamente

```python
T = 5
r = .1
tipo = "P"
#3d plot r y T, contra C

r_vec= np.arange(r*.1,2*r,r/10) #Vectores de datos para r y T
T_vec= np.arange(T*.1,2*T,T/10)

# Generamos el X e Y para plotear. Para generar un plot 3D se tiene que tener 
# NxM puntos de (X,Y), donde evaluar la funcion. Estos vectores generan esa malla 
r_ = np.outer(r_vec, np.ones(len(r_vec)))    
T_ = np.outer(T_vec, np.ones(len(T_vec))).T

#Calculamos los precios de las opciones
precio = np.empty(shape=(len(r_vec),len(T_vec)))

# para cada r hay M valores de T y queremos evaluar en c/u
for i in range(len(r_vec)):
    for j in range(len(T_vec)):
        precio[i,j] = bs(tipo,S,K,T_vec[j],r_vec[i],sigma,q)

#Genera el grafico
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')

#genera la superficie y el gradiente de colores
surf = ax.plot_surface(r_, T_, np.array(precio), cmap = 'hot')
fig.colorbar(surf, ax = ax,
             shrink = 0.5, aspect = 5)

#Titulos
ax.set_title('Precio de Put para S=255 y K=250', fontsize=20)
ax.set_xlabel('Tasa de interes anualizada(r)', fontsize=15)
ax.set_ylabel('Expiracion en años (T)', fontsize=15)

plt.show()
```

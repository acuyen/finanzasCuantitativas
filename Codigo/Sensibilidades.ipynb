## 1) Sensibilidades


```python
### Importo los modulos
from opcion_europea_bs import opcion_europea_bs as bs
import numpy as np
from matplotlib import pyplot as plt
```

## API Yahoo finance


### Parametros iniciales

```python
S = 255
K = 250
T = 3/12
r = .02
q = 0
sigma = 0.5

```

### Precio del Call y del put con estos parametros

```python
tipo = "C"
call = bs(tipo,S,K,T,r,sigma,q)

print("El precio del call es ", call)
```

```python
tipo = "P"
call = bs(tipo,S,K,T,r,sigma,q)

print("El precio del put es ", call)
```

### Sensibilidad en el spot ($S$)

```python
tipo = "C"
S_vec = np.arange(S*.1,2*S,S/10)
precios = np.zeros(len(S_vec))

for i in range(len(S_vec)):
    precios[i] = bs(tipo,S_vec[i],K,T,r,sigma,q)
    ## Last VS (B+A)/2

plt.plot(S_vec, precios,'*')


plt.legend()

plt.xlabel('Spot')
plt.ylabel('Precio de mercado del Call')
plt.title('Precio de mercado de las opciones Call con respecto al Spot')
    
plt.show()
    
```

```python
tipo = "P"
S_vec = np.arange(S*.1,2*S,S/10)
precios = np.zeros(len(S_vec))
for i in range(len(S_vec)):
    precios[i] = bs(tipo,S_vec[i],K,T,r,sigma,q)
    ## Last VS (B+A)/2

plt.plot(S_vec, precios,'*')


plt.legend()

plt.xlabel('Spot')
plt.ylabel('Precio de mercado del Put')
plt.title('Precio de mercado de las opciones Put con respecto al Spot')
    
plt.show()
```
